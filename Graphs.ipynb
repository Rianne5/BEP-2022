```python
table_cos_abs=table.merge(table_cosine_abstract, on='url')
sample = sample.merge(table_cosine_abstract, on='url');
```

```python
sample['month'].map(arg,na_action='ignore')
sample['dt_Anth'] = pd.to_datetime(sample['year'].astype(str)+ ' '+sample['month'].map(arg,na_action='ignore'))
sample['dt_Arxiv'] = pd.to_datetime(sample['published_Arxiv'].apply(lambda s: s[:-15]),utc=False )
sample['dif'] = sample['dt_Anth']-sample['dt_Arxiv']
```

```python
table_cos_abs['month'].map(arg,na_action='ignore')
table_cos_abs['dt_Anth'] = pd.to_datetime(table_cos_abs['year'].astype(str)+ ' '+table_cos_abs['month'].map(arg,na_action='ignore'))
table_cos_abs['dt_Arxiv'] = pd.to_datetime(table_cos_abs['published_Arxiv'].apply(lambda s: s[:-15]),utc=False )
table_cos_abs['dif'] = table_cos_abs['dt_Anth']-table_cos_abs['dt_Arxiv']
```

```python
sample['day'] = sample['dif'].apply( lambda x:x.days)
table_cos_abs['day'] = table_cos_abs['dif'].apply( lambda x:x.days)
```

```python
# table for time statistics in report

time = table[table['title_Arxiv'].isna()==False][table['title_Arxiv']!='NF']
time['month'].map(arg,na_action='ignore')
time['dt_Anth'] = pd.to_datetime(time['year'].astype(str)+ ' '+time['month'].map(arg,na_action='ignore'))
time['dt_Arxiv'] = pd.to_datetime(time['published_Arxiv'].apply(lambda s: s[:-15]),utc=False )
time['dif'] = time['dt_Anth']-time['dt_Arxiv']
time['day'] = time['dif'].apply(lambda x:x.days)
```
